```python
import sqlite3
import pandas as pd
conn = sqlite3.connect('sqlite_db_pythonsqlite.db')
```

```python
cur = conn.cursor()
```

```python
# Q10: Produce a list of facilities with a total revenue less than 1000. 
# The output of facility name and total revenue, sorted by revenue. 
# Rememberthat there's a different cost for guests and members!

res = cur.execute('SELECT * \
                FROM (SELECT Facilities.name, \
                    SUM(CASE WHEN memid = 0 \
                    THEN guestcost * slots \
                    ELSE membercost * slots \
                    END) AS total_revenue \
                FROM Facilities \
                INNER JOIN Bookings ON Facilities.facid = Bookings.facid \
                GROUP BY Facilities.name) AS sub \
                WHERE total_revenue < 1000 \
                ORDER BY total_revenue;')
sol = res.fetchall()
sol
```

```python
# Q11: Produce a report of members and who recommended them in alphabetic surname,firstname order
sql_query11 = pd.read_sql_query("SELECT A.surname || ', ' || A.firstname AS member, B.surname || ', ' || B.firstname AS recommended_by \
                            FROM Members A, Members B \
                            WHERE A.recommendedby = B.memid AND A.memid <> 0 \
                            ORDER BY member, recommended_by;", conn)
df11 = pd.DataFrame(sql_query11)
df11
```

```python
# Q12: Find the facilities with their usage by member, but not guests

sql_query12 = pd.read_sql_query("SELECT b.facid, f.name, COUNT(b.memid) AS 'usage by members' \
                            FROM Bookings as b \
                            LEFT JOIN Facilities as f ON b.facid = f.facid \
                            WHERE memid <> 0 \
                            GROUP BY b.facid, f.name \
                            ORDER BY b.facid;", conn)
df12 = pd.DataFrame(sql_query12)
df12
```

```python
# Q13: Find the facilities usage by month, but not guests
sql_query13 = pd.read_sql_query("SELECT strftime('%m', starttime) AS month, COUNT( memid ) AS 'usage by members' \
                            FROM  Bookings \
                            WHERE memid <> 0 \
                            GROUP BY month;", conn)
df13 = pd.DataFrame(sql_query13)
df13
```
