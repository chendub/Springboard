```python
# imports
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# load data
boulder = pd.read_csv('data/boulder_results.csv')
```

```python
boulder.info()
```

```python
boulder.head()
```

```python
# extracting the location of the competiton from the Competition Title column into a new column 'Location'
boulder['Location'] = boulder['Competition Title'].str.extract(r'\(([A-Z]{3})\)')
boulder.head()
```

```python
# create a new column 'home' that indicates if the athlete is competeing in their home country

boulder['home'] = boulder['Nation'].eq(boulder['Location'])
boulder.head()
```

```python
# check to see how many unique values for Nation
nations_list = boulder['Nation'].unique()
print(nations_list)
print('Climbers are from ', len(nations_list), ' different nations')
```

```python
# How many unique countries where competitions are held?
locations_list = boulder['Location'].unique()
print(locations_list)
print('Competitions are held in', len(locations_list), 'different countries')
```

```python
# How many individual athletes just by first and last name
boulder.groupby(['FIRST','LAST']).size()
```

```python
# How many individual athletes by name and nation
boulder.groupby(['FIRST', 'LAST', 'Nation']).size()
```

```python
# Athletes with same name but different nation?
w_nation = boulder.groupby(['FIRST', 'LAST', 'Nation']).size().to_frame(name = 'count').reset_index()
no_nation = boulder.groupby(['FIRST', 'LAST']).size().to_frame(name = 'count').reset_index()
difference = pd.concat([w_nation.drop('Nation', 1), no_nation]).drop_duplicates(keep=False)
difference
```

```python
# number of athletes per natioanl teams

athletes = w_nation.drop('count', 1).drop_duplicates(keep=False)
athletes.groupby('Nation').size().sort_values(ascending=False)
```

```python
# plotting team sizes
athletes.groupby('Nation').size().sort_values(ascending=False).plot(kind='barh')
plt.rcParams['figure.figsize'] = [6, 10]
plt.title('National Teams Sizes')
plt.xlabel('Count')
plt.subplots_adjust(wspace=0.5)
```
